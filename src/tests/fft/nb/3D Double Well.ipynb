The point is to use a non-periodic potential, but periodic BC. The reference energies are using FFT:
```
           1  -3.1862769621639497     
           2  -3.0868421517330287     
           3   7.6420863657700622E-002
           4  0.49315729619284099     
```
One can check that if a periodic potential is used (by solving a Poisson equation), then one obtains:
```
           1  -3.0377882099791229     
           2  -2.9383534216203397     
           3  0.22490960438627461     
           4  0.64164581389585251     
```
Which are all equivalent, just shifted by `0.14848875218482682` and differ just by `1e-7` or so. As such, we can just use the non-periodic potential directly.
