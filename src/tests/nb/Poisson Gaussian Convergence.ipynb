```python
%matplotlib inline
from math import sqrt, pi
import matplotlib.pyplot as plt
import numpy as np
```

Generate `data.txt` using `test_poisson_fft`.

```python
D = np.loadtxt("../data.txt")
```

```python
Ng = D[:, 0]
alpha = D[:, 1]
E = D[:, 2]
```

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot(111)

for i in range(6, 200, 10):
    ax.semilogy(Ng[alpha==i], abs(E[alpha==i] - E[alpha==i][-1]), "x-", label=r"$\alpha=%d$" % i)
    
# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.xlabel(r"$N_q$ (# of PW in each direction)")
plt.ylabel("Error in Hartree energy [a.u.]")
plt.title(r"FFT convergence of a Gaussian density with various $\alpha$")
plt.show()
```

Only $\alpha \le 66$ are converged to $10^{-6}\rm\,a.u.$ accuracy for $N_q=256$. Use the above graph to assess accuracy for the given $\alpha$ of the following plot:

```python
x = np.array(range(6, 80, 10))
y = np.array([E[alpha==i][-1] for i in x])
plt.figure(figsize=(10, 6))
plt.plot(x, y-x/sqrt(2*pi), "o-")
plt.grid()
plt.xlabel(r"$\alpha$")
plt.ylabel("Hartree energy [a.u.]")
plt.title(r"Convergence of a Gaussian density to a delta function ($\alpha\to\infty$)")
plt.show()
```

```python
E_ewald = -0.70932436987015390
```

```python
x = np.array(range(6, 200, 10))
y = np.array([E[alpha==i][-1] for i in x]) - x/sqrt(2*pi)
plt.figure(figsize=(10, 6))
plt.semilogy(x, abs(y-E_ewald), "o-")
plt.grid()
plt.xlabel(r"$\alpha$")
plt.ylabel("Error in Hartree energy compared to Ewald [a.u.]")
plt.title(r"Convergence of a Gaussian density to a delta function ($\alpha\to\infty$)")
plt.show()
```

It converges to ~ 1e-4 for ~ $\alpha=66$, then it starts diverging, because the accuracy from FFT goes down, per our convergence graph above.
