### Conclusion

The positive nuclei pseudopotential $V(r)$ is defined on $[0, R_{cut}]$ and it becomes ${1\over r}$ on $[R_{cut}, \infty]$, agreeing at least in values and first derivatives.

`H.pseudo` contains uniform radial grid [0, Rcut] (scaled to [0, 1]) and the positive V(r) potential in the two columns.

`RegularizedPotential.dat` contains the non-uniform radial grid $[10^{-8}, 1.86]$ and the functions $-r V(r)$ and $-{Z \over r}$ in the three columns.

`RegularizedDensity.dat` contains another non-uniform radial grid and the regularized potential $n_r(r)$ and non-regularized potential $n(r)$.
