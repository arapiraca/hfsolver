```python
from numpy import (empty, pi, meshgrid, linspace, sum, sin, exp, shape, sqrt,
        conjugate, loadtxt)
from numpy.fft import fftn, fftfreq, ifftn

f = open("Ven_coulomb.txt")
s = array([int(x) for x in f.readline().split()])
D = array([float(x) for x in f.read().split()])
N = s[0]
Ven = D.reshape((N, N, N))
print "N =", N
L = 2.
x1d = linspace(-L/2, L/2, N+1)[:-1]
x, y, z = meshgrid(x1d, x1d, x1d, indexing="ij")
r = sqrt(x**2+y**2+z**2)
#plot(x[:, N/2, N/2], (r**2)[:, N/2, N/2])
Ven_analytic = -1/r
shift = Ven[0, 0, 0] - Ven_analytic[0, 0, 0]
print "shift =", shift
Ven_analytic += shift
z = N/2 # the index corresponding to y=z=0
plot(x[:, z, z], Ven[:, z, z], label="fft")
plot(x[:, z, z], Ven_analytic[:, z, z], label="analytic")
legend(loc="lower right")
title("Comparion of Ven from FFT vs analytic")
xlabel("x [a.u.]")
ylabel("$V_{en}(x, 0, 0)$ [a.u.]")
savefig("Ven.png");
```
