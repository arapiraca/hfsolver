```python
D = loadtxt("../H.pseudo.orig", skiprows=1)
```

```python
R = D[:, 0]
V = D[:, 1]
```

```python
plot(R, V)
plot(R, 1/R)
ylim([1, 7])
```

```python
from scipy.interpolate import splrep, splev
tck = splrep(R, V)
n_spline = -(splev(R, tck, der=2)+2/R*splev(R, tck, der=1))/(4*pi)
```

```python
semilogx(R, n_spline)
```
