```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
```

```python
D = np.loadtxt("../rdirac1.txt")
plt.figure(figsize=(16, 16))
for p in range(1, 11):
    DOFs = D[:, 0]
    p2 = D[:,1]
    Ne = D[:, 2]
    eigs = D[:, 7:]
    condA = D[:, 5]
    condB = D[:, 6]
    E_ref = -489.03708487635595

    x = DOFs[p2==p]
    if (p<10):
        i = 1 # Skip spurious state
    else:
        i = 0
    E0 = eigs[p2==p,i]
    y = abs(E0-E_ref)
    order = 2*p+1
    y0 = 1/x**order
    n = -4
    y0 = y0 / y0[n] * y[n]
    plt.subplot(221)
    plt.loglog(x, y, "o-", label="p=%d" % p)
    #plt.loglog(x, y0, "--", label="order %d" % order)
    plt.subplot(222)
    plt.loglog(x, condB[p2==p], label="B")
    plt.subplot(223)
    plt.loglog(x, condA[p2==p], label="A")

plt.gcf().suptitle("Direct Hamiltonian", fontsize=14)
plt.subplot(221)
plt.xlabel("Degrees of freedom")
plt.ylabel("Error in lowest eigenvalue [a.u.]")
plt.legend()
plt.grid()
plt.subplot(222)
plt.xlabel("Degrees of freedom")
plt.ylabel("Condition number S")
plt.grid()
plt.subplot(223)
plt.xlabel("Degrees of freedom")
plt.ylabel("Condition number H")
plt.grid()
plt.savefig("direct_hamiltonian.pdf")
plt.show()
```

```python
D = np.loadtxt("../rdirac2.txt")
plt.figure(figsize=(16, 16))
for p in range(1, 11):
    DOFs = D[:, 0]
    p2 = D[:,1]
    Ne = D[:, 2]
    eigs = D[:, 7:]
    condA = D[:, 5]
    condB = D[:, 6]
    E_ref = -489.03708487635595

    x = DOFs[p2==p]
    E0 = eigs[p2==p,0]
    y = abs(E0-E_ref)
    order = 2*p
    y0 = 1/x**order
    n = -4
    y0 = y0 / y0[n] * y[n]
    plt.subplot(221)
    plt.loglog(x, y, "o-", label="p=%d" % p)
    #plt.loglog(x, y0, "--", label="order %d" % order)
    plt.subplot(222)
    plt.loglog(x, condB[p2==p], label="B")
    plt.subplot(223)
    plt.loglog(x, condA[p2==p], label="A")

plt.gcf().suptitle("Squared Hamiltonian", fontsize=14)
plt.subplot(221)
plt.xlabel("Degrees of freedom")
plt.ylabel("Error in lowest eigenvalue [a.u.]")
plt.ylim([1e-6, 1e3])
plt.legend()
plt.grid()
plt.subplot(222)
plt.xlabel("Degrees of freedom")
plt.ylabel("Condition number S")
plt.grid()
plt.subplot(223)
plt.xlabel("Degrees of freedom")
plt.ylabel("Condition number H")
plt.grid()
plt.savefig("squared_hamiltonian.pdf")
plt.show()
```
