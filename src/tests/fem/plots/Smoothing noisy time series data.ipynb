## Smoothing noisy time series data

*By Gregory von Winckel*

Suppose we have some time series data $z(t)$ evaluated on $t_k=k\Delta t$ where the data is noisy, but we believe that it should be a smooth function. For simplicity, let $t \in [-1, 1]$. To make a smooth approximation $y$ of the noisy data, we can formulate an unconstrained optimization problem. Let $y$ be the minimizer of the cost functional
$$
J(y) = \frac{1}{2}\|y-z\|^2 + \frac{\alpha}{2}\left\|\frac{d^k y}{dt^k}\right\|^2
$$
Where increasing $k$ makes the solution smoother and decreasing $\alpha$ makes the solution $y$ closer to the data $z$. The first order optimality condition specifies that $y$ satisfy the differential equation
$$
-\alpha\frac{d^{2k}y}{dt^{2k}} +y = z
$$
To make this solveable, we need to impose $2k$ boundary or other conditions that $y$ must satisfy. We impose integral conditions
$$
\int\limits_{-1}^1 y t^j \,\mathrm{d}t = \int\limits_{-1}^1 z t^j \,\mathrm{d}t,\quad j=0,...,2k-1
$$
Which means that the residual of the fit is orthogonal to a polynomial subspace $y-z\bot \mathbb{P}^{2k-1}$ where 
$$
\mathbb{P}^{2k-1} = \text{span}\{1,t,...,t^{2k-1}\}
$$
It turns out that these conditions are very convenient to implement if we discretize the differential equation using a Legendre polyomial basis
$$
y(t)=\sum\limits_{j=0}^m \hat y_j P_j(t)
$$
The integral conditions mean that the first $2k$ Legendre expansion coefficients of $y$ are equal to the first $2k$ expansion coefficients of $z$. Operating under the assumption that we might like to smooth a lot of data on the same grid, we implement the smoothing algorithm as a class which contains member data that will be reused repeatedly.

```python
%pylab inline

import numpy as np
from numpy.polynomial.legendre import legvander,legder
from scipy.linalg import qr,solve_triangular,lu_factor,lu_solve

class legendre_smooth(object):
    def __init__(self,n,k,a,m):
        self.k2 = 2*k

        # Uniform grid points
        x = np.linspace(-1,1,n)

        # Legendre Polynomials on grid 
        self.V = legvander(x,m-1)

        # Do QR factorization of Vandermonde for least squares 
        self.Q,self.R = qr(self.V,mode='economic')

        I = np.eye(m)
        D = np.zeros((m,m))
        D[:-self.k2,:] = legder(I,self.k2)

        # Legendre modal approximation of differential operator
        self.A = I-a*D

        # Store LU factors for repeated solves   
        self.PLU = lu_factor(self.A[:-self.k2,self.k2:])


    def fit(self,z):

        # Project data onto orthogonal basis 
        Qtz = np.dot(self.Q.T,z)

        # Compute expansion coefficients in Legendre basis
        zhat = solve_triangular(self.R,Qtz,lower=False)

        # Solve differential equation       
        yhat = np.zeros(len(zhat))
        q = np.dot(self.A[:-self.k2,:self.k2],zhat[:self.k2])
        r = zhat[:-self.k2]-q
        yhat[:self.k2] = zhat[:self.k2]
        yhat[self.k2:] = lu_solve(self.PLU,r)
        y = np.dot(self.V,yhat)
        return y
```

Without some a priori knowledge of what the best choices of $k$ and $\alpha$ are as well as how many Legendre modes are needed, some experimentation is needed to find a good fit to the data. We can test the performance of the method with

```python
import matplotlib.pyplot as plt

n = 3000
t = np.linspace(-4,4,n)

z = np.tanh(t) 
zn = z + 0.02*np.random.randn(n)

smooth = legendre_smooth(n,12,1e-4,80)

y = smooth.fit(zn)

fig = plt.figure(1, figsize=(15, 10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)


ax1.set_title('Noisy data')
ax2.set_title('Original and smoothed noisy data')
ax3.set_title('Errors')


ax1.plot(t, zn)
ax2.plot(t,z,'k',t,y,'r--',lw=2)
ax3.semilogy(t, abs(z-y))
plt.show()
```

### Notes about the idea of minimizing the cost functional

First suppose that we had just the cost functional
$$
J(y)=\frac{1}{2}\|y-z\|^2=\frac{1}{2}\int\limits_{-1}^1 (y-z)^2\,\mathrm{d}t
$$
The $y$ which minimizes this equal to $z$ almost everywhere if $t$ is continuous and everywhere is $t$ is discrete. This obviously doesn't help us at all as we want $y$ to be smoother than $z$. A measure of smoothness is how many derivatives are bounded or how small the norms of the derivatives are. By adding a penalty on the derivatives, we can make the minimizer smoother. Supposing we add a penalty on the derivative
$$
J(y)=\frac{1}{2}\|y-z\|^2+\frac{\alpha}{2}\left\|\frac{d^k y}{dt^k}\right\|^2
$$
For this functional to be finite, $y$ must have $k$ bounded derivatives. 
Taking variations with respect to $y$ and setting them equal to zero
$$
(J_y,\nu) = \frac{d}{d\epsilon}J(y+\epsilon\nu)\big|_{\epsilon=0}= \int\limits_{-1}^1 (y-z)\nu+\alpha\left(\frac{d^ky}{dt^k}\right) \left(\frac{d^k\nu}{dt^k}\right)\,\mathrm{d}t = 0
$$
We can then write
$$
(\nu,y)+\alpha\left(\frac{d^k \nu}{dt^k},\frac{d^k y}{dt^k}\right)=(\nu,z)
$$
The Babuska-Lax-Milgram theorem says that this has a unique solution where $\nu \in V$ and 
$y\in V^\ast$. Furthermore, since the data $z$ is bounded and defined on a finite interval, it must be an element of $L^2$ and since $(\nu, z)$ is finite for all $\nu$, it must be that $\nu$ is also an $L^2$ function. This means that 
$$
\frac{d^k \nu}{dt^k}\in H^{-k} =: V
$$
Therefore the dual space is $V^\ast = H^k$ so the solution to the weak form of the differential equation $y(t)$ has to be an element of $H^k$, which means it is quite smooth if $k$ is large.
