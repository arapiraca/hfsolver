# Matrix visualization

The matrix is saved in Fortran using:
```fortran
open(newunit=u, file="matrixA.txt", status="replace")
write(u, *) Ap
write(u, *) Aj
write(u, *) Ax
close(u)
```

````python
%pylab inline
from scipy.sparse import csr_matrix
````

````python
f = open("../matrixA.txt")
Ap = fromstring(f.readline(), dtype="int", sep=" ")
Aj = fromstring(f.readline(), dtype="int", sep=" ")
Ax = fromstring(f.readline(), dtype="double", sep=" ")
````

````python
A = csr_matrix((Ax, Aj-1, Ap-1)).toarray()
````

````python
from matplotlib.colors import ListedColormap

def get_hue(x):
    H = 240+180.0*x
    while H >= 360.0:
        H -= 360.0
    while H < 0.0:
        H += 360.0
    return H

def get_map(N):
    V = 1   # The value.
    S = 1   # The saturation.
    C = V*S # The chroma.
    
    colorValues = []
    
    for i in range(N):
        H = get_hue(i/float(N))
        Hprime = H/60.0
        X = C*(1-math.fabs(Hprime%2-1))
        if 0 <= Hprime < 1:
            RGB = [C, X, 0]
        elif 1 <= Hprime < 2:
            RGB = [X, C, 0]
        elif 2 <= Hprime < 3:
            RGB = [0, C, X]
        elif 3 <= Hprime < 4:
            RGB = [0, X, C]
        elif 4 <= Hprime < 5:
            RGB = [X, 0, C]
        elif 5 <= Hprime < 6:
            RGB = [C, 0, X]
        
        colorValues.append("#%02x%02x%02x" % (
            int(math.floor(255*RGB[0])),
            int(math.floor(255*RGB[1])),
            int(math.floor(255*RGB[2]))
            ))
    
    return colorValues

def showmat(A):
    Arelat = abs(A)/abs(A).max()
    figure(figsize=(16,8))
    subplot(121)
    N = 12
    colors = get_map(N+1)
    cmap = ListedColormap(colors[1:])
    im = imshow(Arelat, cmap=cmap, interpolation='nearest', vmin=0, vmax=1)
    title(r"Matrix $A$ (linear scale)")
    colorbar()
    subplot(122)
    Arelatlog = Arelat.copy()
    mask = ~(Arelat == 0)
    Arelatlog[mask] = log(Arelat[mask])
    Arelatlog[~mask] = -13 # less than vmin
    cmap.set_under(colors[0])
    im = imshow(Arelatlog, cmap=cmap, interpolation='nearest', vmin=-12, vmax=0)
    title("Matrix $A$ (log scale)")
    colorbar(extend="min")    
    show()
````

### Matrix $A$

The matrix $A$ in the linear system:
$$
A x = b
$$

````python
showmat(A)
````

### Matrix $A^{-1}$

Let's calculate SVD first:

````python
U, s, V = svd(A)
````

Let's verify that $U S V = A$ and that $U^{-1} = U^T$ and $V^{-1} = V^T$:

````python
print abs(dot(U, dot(diag(s), V))-A).max()
print abs(inv(U)-U.T).max()
print abs(inv(V)-V.T).max()
````

Let's look at the singular values, you can see that the last one is zero:

````python
s
````

So the matrix $A$ can't be inverted. But we can calculate the pseudoinverse by inverting all the singular values and leaving the zero be:

````python
s_inv = 1/s
s_inv[-1] = 0
s_inv
````

````python
Apinv = dot(V.T, dot(diag(s_inv), U.T))
````

The pseudoinverse is "almost" an inverse --- when multiplied by $A$, the result is close to 1 at diagonal, and close to 0 everywhere else:

````python
showmat(dot(Apinv, A))
````

Here is how the pseudoinverse looks like:

````python
showmat(Apinv)
````
