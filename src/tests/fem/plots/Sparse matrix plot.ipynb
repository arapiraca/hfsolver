# Matrix visualization

The matrix is saved in Fortran using:
```fortran
open(newunit=u, file="matrixA.txt", status="replace")
write(u, *) Ap
write(u, *) Aj
write(u, *) Ax
close(u)
```

````python
%pylab inline
from scipy.sparse import csr_matrix
````

````python
f = open("../matrixA.txt")
Ap = fromstring(f.readline(), dtype="int", sep=" ")
Aj = fromstring(f.readline(), dtype="int", sep=" ")
Ax = fromstring(f.readline(), dtype="double", sep=" ")
````

````python
A = csr_matrix((Ax, Aj-1, Ap-1)).toarray()
````

````python
from matplotlib.colors import ListedColormap
from numpy import ma

def get_hue(x):
    H = 240+180.0*x
    while H >= 360.0:
        H -= 360.0
    while H < 0.0:
        H += 360.0
    return H

def get_map(N):
    V = 1   # The value.
    S = 1   # The saturation.
    C = V*S # The chroma.
    
    colorValues = []
    
    for i in range(N):
        H = get_hue(i/float(N))
        Hprime = H/60.0
        X = C*(1-math.fabs(Hprime%2-1))
        if 0 <= Hprime < 1:
            RGB = [C, X, 0]
        elif 1 <= Hprime < 2:
            RGB = [X, C, 0]
        elif 2 <= Hprime < 3:
            RGB = [0, C, X]
        elif 3 <= Hprime < 4:
            RGB = [0, X, C]
        elif 4 <= Hprime < 5:
            RGB = [X, 0, C]
        elif 5 <= Hprime < 6:
            RGB = [C, 0, X]
        
        colorValues.append("#%02x%02x%02x" % (
            int(math.floor(255*RGB[0])),
            int(math.floor(255*RGB[1])),
            int(math.floor(255*RGB[2]))
            ))
    
    return colorValues

def showmat(A):
    Arelat = abs(A)/abs(A).max()
    figure(figsize=(16,8))
    subplot(121)
    N = 512
    colors = get_map(N+1)
    cmap = ListedColormap(colors[1:])
    im = imshow(Arelat, cmap=cmap, interpolation='nearest', vmin=0, vmax=1)
    title(r"Matrix $A$ (linear scale)")
    colorbar()
    subplot(122)
    mask = (Arelat == 0)
    Arelatlog = ma.array(Arelat, mask=mask)
    Arelatlog = ma.log(Arelatlog)
    Arelatlog[mask] = -13 # less than vmin
    cmap.set_under(colors[0])
    im = imshow(Arelatlog, cmap=cmap, interpolation='nearest', vmin=-12, vmax=0)
    title("Matrix $A$ (log scale)")
    colorbar(extend="min")    
    show()
````

### Matrix $A$

The matrix $A$ in the linear system:
$$
A x = b
$$

````python
showmat(A)
````

### Matrix $A^{-1}$

Let's calculate SVD first:

````python
U, s, V = svd(A)
````

Let's verify that $U S V = A$ and that $U^{-1} = U^T$ and $V^{-1} = V^T$:

````python
print abs(dot(U, dot(diag(s), V))-A).max()
print abs(inv(U)-U.T).max()
print abs(inv(V)-V.T).max()
````

Let's look at the singular values, you can see that the last one is zero:

````python
s[-20:]
````

So the matrix $A$ can't be inverted. But we can calculate the pseudoinverse by inverting all the singular values and leaving the zero be:

````python
s_inv = 1/s
s_inv[-1] = 0
s_inv[-20:]
````

````python
Apinv = dot(V.T, dot(diag(s_inv), U.T))
````

The pseudoinverse is "almost" an inverse --- when multiplied by $A$, the result is close to 1 at diagonal, and close to 0 everywhere else:

````python
showmat(dot(Apinv, A))
````

Here is how the pseudoinverse looks like:

````python
showmat(Apinv)
````

### How to use pseudoinverse to solve the linear system

Let's first create some random right hand side $b$ and make sure it is net neutral:

````python
n = shape(A)[0]
b0 = rand(n)
b = b0 - sum(b0) / n
print "sum(b) =", sum(b)
````

Solve the linear system $Ax = b$ directly and save to $x_1$:

````python
x1 = solve(A, b)
````

Solve the linear system $Ax = b$ by using the pseudoinverse $x_2 = A^{-1} b$:

````python
x2 = dot(Apinv, b)
````

Let's check that both solutions solve the original problem:

````python
print "residual 1 =", norm(dot(A, x1)-b)
print "residual 2 =", norm(dot(A, x2)-b)
````

Note that we can add an arbitrary constant to the solution:

````python
print norm(dot(A, x1+5)-b)
print norm(dot(A, x1+13)-b)
````

The two solutions $x_1$ and $x_2$ differ by a constant $C$:

````python
C = x2[0] - x1[0]
C
````

But otherwise they are exactly equal:

````python
norm(x2-(x1+C))
````

The null space of A is given by the last column $x_0$ of $V$:

````python
x0 = V[-1, :]
````

i.e. $Ax_0 = 0$, so $x_0$ spans the null space:

````python
norm(dot(A, x0))
````

It's interesting that $x_0$ is a constant vector:

````python
C = x0[0]
print C
print norm(x0-C)
````

Then $t x_0$ spans the null space, so if $x$ is a solution, then $x+tx_0$ is also a solution for any (scalar) $t$.
But because $x_0$ is a constant vector, it means that we can normalize it to all ones and $x+tx_0$ becomes just $x+t$. In other words, we can add a constant to any solution and it will still be a solution, as we observed above.

**NOTE**: If the right hand side is not net-neutral, then the system can't be solved, as shown here by using the original not net-neutral $b_0$:

````python
x1 = solve(A, b0)
x2 = dot(Apinv, b0)
````

Then the solution does not exist:

````python
print "residual 1 =", norm(dot(A, x1)-b0)
print "residual 2 =", norm(dot(A, x2)-b0)
````

Also the two solutions differ by a large number:

````python
C = x2[0] - x1[0]
"%e" % C
````

````python
norm(x2-(x1+C))
````
