# Matrix visualization

The matrix is saved in Fortran using:
```fortran
open(newunit=u, file="matrixA.txt", status="replace")
write(u, *) matAp
write(u, *) matAj
write(u, *) matAx
close(u)
```

````python
%pylab inline
from scipy.sparse import csr_matrix
````

````python
f = open("../matrixA.txt")
Ap = fromstring(f.readline(), dtype="int", sep=" ")
Aj = fromstring(f.readline(), dtype="int", sep=" ")
Ax = fromstring(f.readline(), dtype="double", sep=" ")
````

````python
A = csr_matrix((Ax, Aj-1, Ap-1))
````

### Matrix $A$

The matrix $A$ in the linear system:
$$
A x = b
$$

````python
cmap = cm.pink_r
im = imshow(abs(A.todense()), cmap=cmap, interpolation='nearest')
title("Matrix $A$ (linear scale)")
colorbar();
````

````python
cmap = cm.pink_r
im = imshow(log(abs(A.todense())), cmap=cmap, interpolation='nearest')
title("Matrix $A$ (log scale)")
colorbar();
````

````python
cmap = cm.pink_r
im = imshow(log(abs(A.todense())/abs(A.todense()).max()), cmap=cmap, interpolation='nearest')
title("Matrix $A$ (log scale)")
colorbar();
````

### Matrix $A^{-1}$

For the inverse matrix we get:

````python
Ainv = inv(A.todense())
````

````python
cmap = cm.pink_r
im = imshow(abs(Ainv), cmap=cmap, interpolation='nearest')
title("Matrix $A^{-1}$ (linear scale)")
colorbar();
````

````python
cmap = cm.pink_r
im = imshow(log(abs(Ainv)/abs(Ainv).max()), cmap=cmap, interpolation='nearest')
title("Matrix $A^{-1}$ (log scale)")
colorbar();
````

````python
cmap = cm.pink_r
im = imshow(log(abs(Ainv)), cmap=cmap, interpolation='nearest')
title("Matrix $A^{-1}$ (log scale)")
colorbar();
````
