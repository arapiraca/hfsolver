## H.pseudo

```python
Rcut = 0.5578590554
D = loadtxt("../H.pseudo.orig", skiprows=1)
R = D[:, 0]*Rcut
V = D[:, 1]
Rout = linspace(R[-1], 1, 30)
R = concatenate((R[:-1], Rout))
V = concatenate((V[:-1], 1/Rout))
```

```python
plot(R, V)
```

### Splines

Calculate cubic splines interpolation.

```python
from scipy.interpolate import splrep, splev
tck = splrep(R, V)
n_spline = -(splev(R, tck, der=2)+2/R*splev(R, tck, der=1))/(4*pi)
```

## Radial Fourier Transform

```python
from scipy.integrate import simps
```

```python
simps(V, R)
```

```python
def radial_fourier_omega(R, f, omega, Z=1.):
    Rcut = R[-1]
    return 4*pi/omega * (simps(R*sin(omega*R)*f, R) + Z*cos(omega*Rcut)/omega)

def radial_fourier_omega_inv(omega, f, R):
    return 1/(2*pi**2*R) * simps(omega*sin(omega*R)*f, omega)

def radial_fourier(R, f, omega_vec):
    return array([radial_fourier_omega(R, f, omega) for omega in omega_vec])

def radial_fourier_inv(omega, f, R_vec):
    return array([radial_fourier_omega_inv(omega, f, R) for R in R_vec])
```

```python
omega = linspace(1e-5, 40, 1000)
Vhat = radial_fourier(R, V, omega)
```

```python
semilogy(omega, abs(Vhat))
```

```python
Vreal = radial_fourier_inv(omega, Vhat, R[1:])
```

```python
plot(R, V, label="orig")
plot(R[1:], Vreal, label="fft back/forth")
legend();
```

```python
plot(R[1:], V[1:]-Vreal)
```

```python
Vderiv = radial_fourier_inv(omega, omega**2*Vhat/(4*pi), R[1:])
```

```python
figure(figsize=(12, 10))
plot(R[1:], Vderiv, label="fft deriv")
plot(R, n_spline, label="spline deriv")
legend();
```
