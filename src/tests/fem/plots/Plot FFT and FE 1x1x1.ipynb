```python
from pylab import plot, semilogy
from numpy import loadtxt, pi, linspace
D = loadtxt("conv.txt")
N = D[:, 0]
DOFs = N**3
E = D[:, 1]
E_exact = D[-1, 1]

calc = abs(E-E_exact)
k = 2.8
A = calc[0] / exp(-k*DOFs[0]**(1./3))
print "E_exact = %.17f" % E_exact
print "A =", A
pred_x = linspace(DOFs[0], DOFs[6], 500)
pred_y = A * exp(-k*pred_x**(1./3))

D1 = loadtxt("conv_fe1.txt")
DOFs_fe1 = D1[:, 1]
E_fe1 = D1[:, 2]

D2 = loadtxt("conv_fe1Nq.txt")
DOFs_fe2 = D2[:, 1]
E_fe2 = D2[:, 2]

calc2 = abs(E_fe2 - E_exact)
k2 = 1.4
A2 = calc2[0] / exp(-k2*DOFs_fe2[0]**(1./3))
print "A2 =", A2
pred_x2 = linspace(DOFs_fe2[0], DOFs_fe2[-1], 500)
pred_y2 = A2 * exp(-k2*pred_x2**(1./3))


D3 = loadtxt("conv_fe3.txt")
DOFs_fe3 = D3[:, 1]
E_fe3 = D3[:, 2]
```

```python
figure(figsize=(8, 6), dpi=80)
semilogy(DOFs[:10]**(1./3), calc[:10], "bo", label="FFT")
semilogy(pred_x**(1./3), pred_y, "g-", label=r"$740.14\cdot\exp(-2.8\cdot \mathrm{DOF}^{1/3})$")
semilogy(DOFs_fe1**(1./3), abs(E_fe1 - E_exact), "ro-", label="FE 1x1x1 $N_q = p+1$")
semilogy(DOFs_fe2**(1./3), abs(E_fe2 - E_exact), "mo-", label="FE 1x1x1 $N_q = 2p+1$")
semilogy(pred_x2**(1./3), pred_y2, "k-", label=r"$23.23\cdot\exp(-1.4\cdot \mathrm{DOF}^{1/3})$")
#semilogy(DOFs_fe3, abs(E_fe3 - E_exact), "mo-", label="FE 3x3x3")
grid()
title("Convergence of an FFT Poisson solver (log-linear plot)")
xlabel(r"n (number of PW in each dimension); p (polynomial order); $\mathrm{DOF}^{1/3}$")
ylabel("abs(E  -  E_exact)  [a.u.]")
#ylim([1e-16, 5e3])
#legend(prop={"size": 13})
legend()
savefig("convergence_log-linear.png")
```
