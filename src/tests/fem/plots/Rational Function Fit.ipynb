```python
from scipy.optimize import curve_fit
```

## Quadratic / Quadratic Fit

```python
def f(x, A0, A1, A2, B1, B2):
    return (A0+A1*x+A2*x**2) / (1+B1*x+B2*x**2)
p0 =(-3.005450, 0.368829, -0.006828, -0.011234, -0.000306)
opt, pcov = curve_fit(f, temp, tec, p0=p0)
print "coefs: ", opt
print "stddev:", sqrt(diag(pcov))
A0, A1, A2, B1, B2 = opt
```

```python
plot(temp, tec, "kx")
x = linspace(min(temp), max(temp), 1000)
plot(x, f(x, A0, A1, A2, B1, B2), "k-")
xlabel("Temperature [K]")
ylabel("Coefficient of thermal expansion")
title("Thermal expansion of Copper data\nQ/Q Rational Function Model");
```

```python
plot(temp, tec-f(temp, A0, A1, A2, B1, B2), "kx")
xlabel("Temperature [K]")
ylabel("Coefficient of thermal expansion")
title("Residuals from Q/Q Fit");
```

## Cubic / Cubic Fit

```python
def f(x, A0, A1, A2, A3, B1, B2, B3):
    return (A0+A1*x+A2*x**2+A3*x**3) / (1+B1*x+B2*x**2+B3*x**3)
p0 = (-2.323648e+00, 3.530298e-01, -1.383334e-02, 1.766845e-04,
      -3.395949e-02, 1.100686e-04, 7.910518e-06)
opt, pcov = curve_fit(f, temp, tec, p0=p0)
print "coefs: ", opt
print "stddev:", sqrt(diag(pcov))
A0, A1, A2, A3, B1, B2, B3 = opt
```

```python
plot(temp, tec, "kx")
x = linspace(min(temp), max(temp), 1000)
plot(x, f(x, A0, A1, A2, A3, B1, B2, B3), "k-")
xlabel("Temperature [K]")
ylabel("Coefficient of thermal expansion")
title("Thermal expansion of Copper data\nC/C Rational Function Model");
```
