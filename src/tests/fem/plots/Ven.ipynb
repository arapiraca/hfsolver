```python
from numpy import linspace
x = linspace(-1, 1, 100)
L = 2
R = sqrt(x**2)
V = -1/R
Ven = 0
N = 10000
for i in range(-N, N+1):
    Ven += -1/abs(x+i*L)
plot(x, V)
plot(x, -1/abs(x-L))
plot(x, -1/R -1/abs(x-L))
plot(x, -1/R -1/abs(x-L)-1/abs(x-2*L))
plot(x, -1/R -1/abs(x-L)-1/abs(x-2*L)-1/abs(x-3*L))
plot(x, Ven, "k-")
```
