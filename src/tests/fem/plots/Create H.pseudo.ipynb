Create the density and the corresponding potential

```python
from numpy import linspace
from scipy.special import erf
Rcut = 0.5496987445
N = 10000
R = linspace(1./N, 1, N)*Rcut
Z = 1.
alpha = 12
V = Z*erf(alpha*R)/R
n = Z*alpha**3/pi**(3./2)*exp(-alpha**2*R**2)
```

Plot it

```python
plot(R, V, "k-", lw=2, label="$-V_{en}(r)$")
plot(R, 1/R, "r--", lw=2, label="-1/R")
plot([Rcut, Rcut], [-100, 0], label="Rcut")
xlim([0, 0.6])
ylim([0, 20])
title("Plot $V_{en}$")
legend()
xlabel("$R$ [a.u.]")
ylabel("potential [a.u.]")
show()
clf()
plot(R, n, "k-", lw=2, label="n(r)")
#plot(R, -1/R, "r--", lw=2, label="-1/R")
plot([Rcut, Rcut], [0, 500], label="Rcut")
xlim([0, 0.6])
ylim([-20, 350])
title("Plot $n_{en}$")
legend(loc="upper right")
xlabel("$R$ [a.u.]")
ylabel("density [a.u.]")
show()
```

Save to file

```python
savetxt("H.pseudo", array([R/Rcut, V]).T, header="%f %d %f %f" % (Z, size(R), Rcut, -0.2845146076), comments="")
```

### Cubic Spline Interpolation

The radial Poisson equation is:
$${1\over r}\left(rV(r)\right)'' = V''(r) + {2\over r} V'(r) = -4\pi n(r)$$
so we calculate $n(r)$ using:
$$n(r) = -{1\over4\pi}\left(V''(r) + {2\over r} V'(r)\right)$$
where $V'(r)$ and $V''(r)$ is calculated from cubic spline interpolation of $V(r)$.

```python
from scipy.interpolate import splrep, splev
tck = splrep(R, V)
n_spline = -(splev(R, tck, der=2)+2/R*splev(R, tck, der=1))/(4*pi)
```

The interpolated $n(r)$ nicely agrees with the original density:

```python
plot(R, n_spline, label="spline")
plot(R, n, label="orig")
xlabel("R")
ylabel("$n(r)$")
title("Nuclear charge density comparison")
legend();
```

```python
semilogy(R, abs(n-n_spline))
xlabel("R")
ylabel("$|n(r) - n_{interp}(r)|$")
title("Absolute error");
```
