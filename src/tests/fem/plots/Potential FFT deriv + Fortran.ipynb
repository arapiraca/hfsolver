## Radial Fourier Transform

```python
from scipy.integrate import simps
```

```python
simps(V, R, even="first")
```

```python
def radial_fourier_omega(R, f, omega, Z=1.):
    Rcut = R[-1]
    return 4*pi/omega * (simps(R*sin(omega*R)*f, R, even="first") + Z*cos(omega*Rcut)/omega)

def radial_fourier_omega_inv(omega, f, R):
    return 1/(2*pi**2*R) * simps(omega*sin(omega*R)*f, omega, even="first")

def radial_fourier(R, f, omega_vec):
    return array([radial_fourier_omega(R, f, omega) for omega in omega_vec])

def radial_fourier_inv(omega, f, R_vec):
    return array([radial_fourier_omega_inv(omega, f, R) for R in R_vec])
```

```python
L = 2.997672536043746
dk = 2*pi/L
omega = linspace(1e-5, 15, 1000)*dk
Vhat = radial_fourier(R, V, omega)
```
