```python
from pylab import plot, semilogy
from numpy import loadtxt, pi
D = loadtxt("conv_fem.txt")
N = D[:, 0]
E = D[:, 1]
DOFs = D[:, 2]
E_exact = E[-1]
print E_exact
```

```python
figure(figsize=(8, 6), dpi=80)
semilogy(DOFs, abs(E-E_exact), "ko")
grid()
title("Convergence of an FE Poisson solver (log-linear)")
xlabel("DOFs (total degrees of freedom)")
ylabel("E  -  E_exact  [a.u.]")
savefig("fe_convergence_semilogy.png")
```

```python
figure(figsize=(8, 6), dpi=80)
calculated = abs(E-E_exact)
loglog(DOFs, calculated, "ko", label="calculated")
predicted = 1/DOFs**6
predicted = predicted * (calculated[-2] /  predicted[-2])
loglog(DOFs, predicted, "g-", label="$1/N^6$")
grid()
title("Convergence of an FE Poisson solver (loglog)")
xlabel("DOFs (total degrees of freedom)")
ylabel("E  -  E_exact  [a.u.]")
legend()
savefig("fe_convergence_loglog.png")
```
