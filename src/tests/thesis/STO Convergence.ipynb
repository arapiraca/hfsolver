```python
from numpy.linalg import cond
from scipy import factorial
from scipy.integrate import simps

def meshexp(rmin, rmax, a, N):
    beta = log(a) / (N-1)
    alpha = (rmax - rmin) / (exp(beta*N) - 1)
    i = arange(N+1)
    return alpha * (exp(beta*i) - 1) + rmin

def N_nzeta(n, zeta):
    return sqrt((2*zeta)**(2*n+1)/factorial(2*n))

def integ(n, zeta):
    # Calculates Integral(r**n * exp(-zeta*r), (r, 0, oo))
    return factorial(n) / zeta**(n+1)

def even_tempered_STO(alpha, beta, N):
    k = arange(1, N+1)
    zeta = alpha*beta**k
    return zeta

def project_n_zeta(n, alpha, nl, zl):
    """
    Projects f(r) = r^n * exp(-alpha*r) to the STO basis nl, zl.
    
    Returns the coefficients of the projection with respect to the STO basis.
    """
    N = size(nl)
    assert N == size(zl)
    A = empty((N, N), dtype="double")
    f = empty(N, dtype="double")
    for l in range(N):
        for k in range(N):
            A[l, k] = integ(nl[l] + nl[k], zl[l] + zl[k])
        f[l] = integ(nl[l] + n, zl[l] + alpha)
    phi = solve(A, f)
    return phi

def project_n_zeta_f(func, r, nl, zl):
    """
    Projects func(r) to the STO basis nl, zl.
    
    Returns the coefficients of the projection with respect to the STO basis.
    """
    N = size(nl)
    assert N == size(zl)
    A = empty((N, N), dtype="double")
    f = empty(N, dtype="double")
    for l in range(N):
        for k in range(N):
            A[l, k] = integ(nl[l] + nl[k], zl[l] + zl[k])
        f[l] = simps(func * r**nl[l] * exp(-zl[l]*r), r)
    phi = solve(A, f)
    return phi


def eval_sto(r, c, nl, zl, normalized=True):
    """
    Evaluates the linear combination "c" of STO basis nl, zl on the grid "r".
    If normalized == True, it will used normalized STO,
    otherwise unnormalized.
    """
    assert size(nl) == size(zl)
    P = zeros(size(r), dtype="double")
    for i in range(size(nl)):
        if normalized:
            N = N_nzeta(nl[i], zl[i])
        else:
            N = 1
        P += phi[i] * r**nl[i] * exp(-zl[i]*r) * N
    return P

def alpha_beta_step(alpha, beta, a, b, N):
    """
    Updates alpha and beta using the iteration step with parameters (a, b).
    
    N ... the number of basis functions *after* the update.
    """
    assert a > 0
    assert -1 < b and b < 0
    beta_old = beta
    beta = beta_old**((N/(N-1.))**b)
    alpha = ((beta-1)/(beta_old-1))**a * alpha
    return alpha, beta

def l2_error(f, g, r):
    return sqrt(simps((f-g)**2, r))
```

Even-tempered basis with constant $\alpha$ and $\beta$ as we increase $N$ converges in about 20 iterations, but to a wrong function, as seen here:

```python
# f(r) = r^n * exp(-alpha*r)
alpha = 1.5
n = 1

# Converge:
r = meshexp(1e-4, 1000, 1e6, 1000)
f = r**n * exp(-alpha*r)
semilogx(r, f, "k-", lw=2, label="f(r)")
N0 = 30
for N in range(N0, N0+5):
    nl = ones(N)
    zl = even_tempered_STO(1.4, 1.7, N)
    phi = project_n_zeta(n, alpha, nl, zl)
    P = eval_sto(r, phi, nl, zl, False)
    semilogx(r, P, label="N = %d" % N)
legend();
```

Even-tempered basis converged by proper update of $\alpha$ and $\beta$ converges in about 20 iterations to the correct function $f(r)$:

```python
# f(r) = r^n * exp(-alpha*r)
alpha = 1.5
n = 1

# Converge:
r = meshexp(1e-4, 1000, 1e6, 1000)
f = r**n * exp(-alpha*r)
semilogx(r, f, "k-", lw=2, label="f(r)")
sto_alpha = 1.4
sto_beta = 1.7
print " N     alpha       beta"
for N in range(1, 20):
    nl = ones(N)
    zl = even_tempered_STO(sto_alpha, sto_beta, N)
    phi = project_n_zeta(n, alpha, nl, zl)
    print "%2d %10.4f %10.4f" % (N, sto_alpha, sto_beta)
    #print "         zl:", zl
    #print "        phi:", phi
    P = eval_sto(r, phi, nl, zl, False)
    if N < 2 or N > 15:
        semilogx(r, P, label="N = %d" % N)
    sto_alpha, sto_beta = alpha_beta_step(sto_alpha, sto_beta, 0.5, -0.5, N+1)
legend();
```
