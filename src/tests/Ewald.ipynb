```python
D = loadtxt("conv3.txt")
n = D[:,0]
E = D[:,1]
D = loadtxt("conv_gauss.txt")
ng = D[:,0]
Eg = D[:,1]
```

```python
err = abs(E-E[-1])
err_est = 1/n**5
err_est *= err[7]/err_est[7]
loglog(n, err_est, lw=2, label="const$\cdot1/n^5$")
loglog(n, err, "o", label="Calculation")
err = abs(Eg-Eg[-1])
err_est = 1/ng**8
err_est *= err[6]/err_est[6]
loglog(ng, err_est, lw=2, label="const$\cdot1/n^8$")
loglog(ng, err, "o", label="Calculation")
grid()
title("FFT convergence to E=%.12f" % E[-1])
legend()
xlabel("Number of plane waves in each direction")
ylabel("Error in Coulomb energy per atom (Ha)")
savefig("coulomb_fft_conv.pdf")
```

```python
D = loadtxt("conv_gauss.txt")
n = D[:,0]
E = D[:,1]
```

```python
err = abs(E-E[-1])
err_est = 1/n**8
err_est *= err[6]/err_est[6]
loglog(n, err_est, lw=2, label="const$\cdot1/n^8$")
loglog(n, err, "o", label="Calculation")
grid()
title("FFT convergence to E=%.12f" % E[-1])
legend()
xlabel("Number of plane waves in each direction")
ylabel("Error in Coulomb energy per atom (Ha)")
savefig("coulomb_fft_conv_gauss.pdf")
```
