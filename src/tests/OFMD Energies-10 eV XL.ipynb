```python
D = loadtxt("Thermodynamics10eV.dat", skiprows=1)
D2 = loadtxt("ofmd_results10eV.txt", skiprows=1)
D3 = loadtxt("ofmd_results10eV_XL.txt", skiprows=1)
```

```python
t  = D[1:, 0]
Fe = D[1:, 1]
Unn = D[1:, 2]
K = D[1:, 3]
F = D[1:, 4]
P = D[1:, 5]
F = Fe + Unn + K
```

```python
t2  = D2[1:, 0]
Fe2 = D2[1:, 1]
Unn2 = D2[1:, 2]
K2 = D2[1:, 3]
F2 = D2[1:, 4]
T2 = D2[1:, 5]
F2 = Fe2 + Unn2 + K2
```

```python
t3  = D3[1:, 0]
Fe3 = D3[1:, 1]
Unn3 = D3[1:, 2]
K3 = D3[1:, 3]
F3 = D3[1:, 4]
T3 = D3[1:, 5]
F3 = Fe3 + Unn3 + K3
```

```python
figure(figsize(16, 8))
plot(t, Fe, "b", lw=2, label="Fe P")
plot(t, Unn, "c", lw=2, label="Unn P")
plot(t, K, "m", lw=2, label="K P")
plot(t, F, "r", lw=2, label="F P")

plot(t2, Fe2, "k--", lw=2, label="Fe")
plot(t2, Unn2, "k--", lw=2, label="Unn")
plot(t2, K2, "k--", lw=2, label="K")
plot(t2, F2, "k--", lw=2, label="F")

plot(t3, Fe3, "k-", lw=1, label="Fe XL")
plot(t3, Unn3, "k-", lw=1, label="Unn XL")
plot(t3, K3, "k-", lw=1, label="K XL")
plot(t3, F3, "k-", lw=1, label="F XL")

xlim([0, 60000])
legend()
grid()
title("Energies")
xlabel("Time [a.u.]")
ylabel("E [eV]")
savefig("MD_energies.pdf")
show()
```

```python
#figure(figsize(16, 8))
plot(t, F - F.mean(), lw=2, label="F")
plot(t, Fe-Fe.mean(), label="Fe")
plot(t, Unn-Unn.mean(), label="Unn")
plot(t, K-K.mean(), label="K")
#plot(t, P-P.mean(), label="P")
grid()
title("Oscillations in Energies")
xlabel("Time [a.u.]")
ylabel("Energy difference from mean [eV]")
#ylim([-2, 2])
#xlim([0, 1600])
legend()
show()
```

```python
figure(figsize(16, 8))
plot(t, F, "r", lw=2, label="F P")
plot(t2, F2, "g-", lw=1, label="F")
plot(t3, F3, "k-", lw=1, label="F XL")
legend()
grid()
title("Free Energy")
xlabel("Time [a.u.]")
ylabel("E [eV]")
xlim([0, 60000])
ylim([-3953.8, -3952.7])

savefig("MD_free_energy.pdf")
show()
```
