```python
D = loadtxt("../ppum_conv.txt")
Nb = D[:, 0]
condA = D[:, 1]
condB = D[:, 2]
eigs = D[:, 3:]
eigs_ref = array([0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
figure(figsize=(8,6))
for i in range(6):
    loglog(Nb, abs(eigs[:,i] - eigs_ref[i]), "o-")

grid()
xlabel("DOFs")
ylabel("Eigenvalue error")
    
y = 1/Nb**6
i = 0 # eig
j = -1 # x-point
y = y / y[j] * abs(eigs[j,i] - eigs_ref[i])
loglog(Nb, y)

#savefig("ppum_fem_conv.pdf")
show()
```
