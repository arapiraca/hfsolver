```python
%pylab inline
from scipy import stats
```

```python
D = loadtxt("../md_results.txt")
```

```python
steps = D.shape[0]
N = (D.shape[1]-4)/(3*3)
t = D[:, 0]
f = D[:, 1:3*N+1]
V = D[:, 3*N+1:6*N+1]
X = D[:, 6*N+1:9*N+1]
Ekin = D[:, 9*N+1]
Epot = D[:, 9*N+2]
Etot = Ekin+Epot
T = D[:, 9*N+3]
f = f.reshape((steps, N, 3))
V = V.reshape((steps, N, 3))
X = X.reshape((steps, N, 3))
```

# General Statistics

```python
N0 = 100
Ekin_avg = Ekin[N0:].mean()
Epot_avg = Epot[N0:].mean()
Etot_avg = Etot[N0:].mean()
figure(figsize=(14, 8))
subplot(2, 2, 1)
plot(t[N0:]*1e12, Ekin[N0:]-Ekin_avg, "-", label="Ekin")
plot(t[N0:]*1e12, Epot[N0:]-Epot_avg, "-", label="Epot")
plot(t[N0:]*1e12, Etot[N0:]-Etot_avg, "-", label="Etot")
legend()
xlabel("Time [ps]")
ylabel("Error")
title("Energy difference from average")

subplot(2, 2, 2)
semilogy(t*1e12, abs(Etot-Etot_avg), "-", label="Etot")
grid()
axvline(t[N0]*1e12, lw=2, color="g")
legend()
xlabel("Time [ps]")
ylabel("abs(Etot-Etot_avg)")
title("Equilibration")


subplot(2, 2, 3)
plot(t[N0:]*1e12, Etot[N0:], "-", label="Etot")
slope, intercept, r_value, p_value, std_err = stats.linregress(t[N0:], Etot[N0:])
plot(t[N0:]*1e12, t[N0:]*slope+intercept, "r-", label="Linear fit")
legend()
xlabel("Time [ps]")
ylabel("Etot")
title("Total Energy")

subplot(2, 2, 4)
plot(t[N0:]*1e12, T[N0:], "x-", label="Temperature")
slope, intercept, r_value, p_value, std_err = stats.linregress(t[N0:], T[N0:])
Tavg = T[N0:].mean()
plot(t[N0:]*1e12, t[N0:]*slope+intercept, "r-", label="Linear fit")
title("Temperature")
xlabel("Time [ps]")
ylabel("E [K]")
legend()

show()
print "T = %f K" % Tavg
```

Temperature cumulative average:

```python
plot(t[N0:], cumsum(T[N0:])/arange(1,steps-N0+1), "x-")
axvline(t[100])
```

Velocity profile at a random time step:

```python
t_step = 500
n, bins, patches = hist(V[t_step, :, 0], 50, normed=1, histtype='stepfilled')
setp(patches, 'facecolor', 'g', 'alpha', 0.75, label="Velocities")
mu = 0
K2au = 3.166811383057024e-06
u2au = 1822.8884848053776
sigma = sqrt(Tavg*K2au / (39.948*u2au))
y = normpdf(bins, mu, sigma)
l = plot(bins, y, 'k--', linewidth=2, label="Maxwell-Boltzmann")
legend()
title("Velocity profile")
show()
```

# Diffusion and velocity correlation

```python
D = loadtxt("../D.txt")
x = D[:, 0]
y = D[:, 1]
velcorr = D[:, 2]
com = D[:, 3:6]
bohr2ang = 0.529177249
y = y * (bohr2ang*1e-8)**2
figure(figsize=(14, 8))
subplot(2, 2, 1)
plot(x*1e12, y)
xlabel("Time [ps]")
title("Diffusion")

subplot(2, 2, 2)
plot(x*1e12, velcorr)
xlabel("Time [ps]")
title("Velocity autocorrelation")

subplot(2, 2, 3)
plot(x*1e12, com[:, 0], label="X")
plot(x*1e12, com[:, 1], label="Y")
plot(x*1e12, com[:, 2], label="Z")
xlabel("Time [ps]")
title("Center of mass coordinates")
legend()

show()
```

Diffusion fit:

```python
N0 = 100
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x[N0:], y[N0:])
plot(x, y)
plot(x[N0:], x[N0:]*slope+intercept)
print slope/6 * 1e5, intercept, std_err/6*1e5
```
